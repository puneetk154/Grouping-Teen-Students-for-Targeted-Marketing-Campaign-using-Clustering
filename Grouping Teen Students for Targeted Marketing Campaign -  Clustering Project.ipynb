```python
# Get the Mean of 'Age' by 'Gradeyear'
tmc.groupby('gradyear').age.mean()
```

```python
# Replace the Missing Values in 'Age' column with Mean value of different Grade Years
tmc['age'] = tmc.groupby('gradyear')['age'].transform(lambda x:x.fillna(x.mean()))
```

```python
# Rectify the Outliers using Inter-Quantile Range
q1 = tmc.age.quantile(.25)
q3 = tmc.age.quantile(.75)
```

```python
iqr = q3-q1
```

```python
(q1-1.5*iqr)
```

```python
(q3+1.5*iqr)
```

```python
tmc = tmc[(tmc.age > (q1-1.5*iqr)) & (tmc.age < (q3+1.5*iqr))]
```

```python
# Outliers have been treated
tmc['age'].describe()
```

```python
# Standard Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
features = sc.fit_transform(features.values)
```

```python
# Label Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
scaled_features
```

```python
# Build K Means Clustering model
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5, random_state=0)
```

```python
model = kmeans.fit(scaled_features)
```

```python
model
```

```python
# Elbow Method
```

```python
inertia = []
for i in range(1,20):
    kmeans=KMeans(n_clusters=i,init = 'k-means++',n_init=10)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)
```

```python
plt.plot(range(1,20), inertia)
```

```python
# Build K Means Clustering model
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 4, random_state = 0)
```

```python
model = kmeans.fit(scaled_features)
```

```python
scaled_features['cluster'] = kmeans.labels_
```
